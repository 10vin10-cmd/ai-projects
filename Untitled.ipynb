```python
from langchain_litellm import ChatLiteLLM
def initialize_any_model(provider: str, model_name: str):
    """
    Uses LiteLLM to dynamically route to any provider using a prefix string.
    Combines 'provider' and 'model_name' into 'provider/model_name'.
    """
    # Formats to: "openai/gpt-4o-mini" or "anthropic/claude-3-5-sonnet-20240620"
    full_model_string = f"{provider}/{model_name}"
    return ChatLiteLLM(
        model=full_model_string,
        json_mode=True,
        
        # 1. Controls creativity/randomness (0.5 as requested)
        temperature=0.5, 
        
        # 2. Controls maximum response length
        max_tokens=256,  
        
        # 3. Advanced: Passes provider-specific 'min_tokens' to the underlying API
        extra_body={"min_tokens": 130} 
    )
   
```

```python
import sys
import os
#from config import Config

# 1. Modern Text Splitter and Vector Store Imports
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
#from langchain_unstructured.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader 
from langchain.retrievers.multi_query import MultiQueryRetriever
#from chatbot import get_litellm_embedding, text_splitter # Reuse your validated embedding engine

# 2. Modern LiteLLM Embeddings Package Integration
from langchain_litellm import LiteLLMEmbeddings

from langchain_community.document_loaders import PyPDFLoader

# You can use this section to suppress warnings generated by your code:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')


# 2. Initialize your dynamic LiteLLM engine (Replacing ModelInference entirely)
# This binds to f"{Config.LLM_PROVIDER}/{Config.MODEL_ID}" automatically
model = initialize_any_model(
    provider=Config.LLM_PROVIDER,
    model_name=Config.MODEL_ID
)

# Apply generation parameters natively using LangChain's configuration binding properties
model.max_tokens = 400
model.temperature = 0.5

# Core Text Splitter function with modern package tracking
def text_splitter(data, chunk_size, chunk_overlap):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
    )
    chunks = splitter.split_documents(data)
    return chunks

# Modern API Embedding configuration function 
def get_litellm_embedding():
    """
    Leverages cloud-hosted API embeddings to avoid local package compilation.
    Automatically maps to standard endpoints using your configured .env credentials.
    """
    return LiteLLMEmbeddings(model="openai/text-embedding-3-small")

# ----------------------------------------------------
# EXECUTION WORKFLOW
# ----------------------------------------------------

# 1. Load document data from local file asset
loader = TextLoader("companypolicies.txt")
txt_data = loader.load()

# 2. Slice text layout into chunks using your parameter metrics
chunks_txt = text_splitter(txt_data, 200, 20)

# 3. Store the token matrices natively inside your in-memory Chromadb workspace
embedding_engine = get_litellm_embedding()
vectordb = Chroma.from_documents(chunks_txt, embedding_engine)

# Define your testing query target string
query = "What is the policy regarding remote work?"


# ----------------------------------------------------
# RETRIEVAL STRATEGY A: Standard Top-K Selection
# ----------------------------------------------------
print("--- Strategy A: Top-1 Similarity Retrieval ---")
retriever_top_k = vectordb.as_retriever(search_kwargs={"k": 1})
docs_top_k = retriever_top_k.invoke(query)

for i, doc in enumerate(docs_top_k, 1):
    print(f"[Doc {i}] Content: {doc.page_content.strip()}")


# ----------------------------------------------------
# RETRIEVAL STRATEGY B: Relevance Threshold Cutoff
# ----------------------------------------------------
print("\n--- Strategy B: Score Threshold Cutoff (0.4) ---")
retriever_threshold = vectordb.as_retriever(
    search_type="similarity_score_threshold", 
    search_kwargs={"score_threshold": 0.4}
)
docs_threshold = retriever_threshold.invoke(query)

if not docs_threshold:
    print("⚠️ No documents matched the strict 40% similarity threshold metric.")
else:
    for i, doc in enumerate(docs_threshold, 1):
        print(f"[Threshold Doc {i}] Content: {doc.page_content.strip()}")
```
